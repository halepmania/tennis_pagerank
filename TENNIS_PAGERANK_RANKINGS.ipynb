# Using Pagerank for Ranking Tennis Players 

This notebook tries to repeat results from Filippo Radicchi's article "Who is the best player ever? A complex network analysis of the history of professional tennis" which can be found at http://arxiv.org/pdf/1101.4028v2.pdf.

```python
# Importing necessary libraries
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import networkx as nx
import operator

plt.rcParams['figure.figsize'] = (18, 5)
 
import os
from IPython.display import HTML
```

This notebook depends on tennis data released by Jeff Sackman (https://github.com/JeffSackmann). This data was slightly modified and can be found at https://github.com/halepmania/tennis_atp/tree/dev. By the way, Jeff maintains very nice resources for tennis addicts: http://www.tennisabstract.com/ and http://heavytopspin.com/. There is a lot of great stuff to dig in.

```python
# Necessary paths for ATP Data
ATP_DATA_FOLDER = 'D:/TennisRepositories/tennis_atp'
ATP_PLAYERS_FILENAME = os.path.join(ATP_DATA_FOLDER, 'atp_players.csv')
ATP_RANKINGS_FILENAME = os.path.join(ATP_DATA_FOLDER, 'atp_rankings_')
ATP_MATCHES_FILENAME = os.path.join(ATP_DATA_FOLDER, 'atp_matches_')
```

Let's write short function to output ATP Ranking for desired date.

```python
def output_atp_rankings(year, date, max_num_players):
    atp_players = pd.read_csv(ATP_PLAYERS_FILENAME)
    atp_rankings = pd.read_csv(ATP_RANKINGS_FILENAME + str(year) + '.csv')
    table = """<table>
    <tr>
    <th>Ranking</th>
    <th>First Name</th>
    <th>Last Name</th>
    <th>Country</th>
    <th>Points</th>
    </tr>
    """
    atp_rankings_date = atp_rankings[atp_rankings['ranking_date'] == date]
    for ranking in xrange(max_num_players):
        atp_ranking = atp_rankings_date[atp_rankings_date['ranking'] == ranking]
        for (i, row) in atp_ranking.iterrows():
            player = atp_players[atp_players['player_id'] == row['player_id']]
            first_name = player['first_name'].values[0]
            last_name = player['last_name'].values[0]
            country_code = player['country_code'].values[0]
            table_row = """
            <tr>
            <td>""" + str(ranking) + """</td>
            <td>""" + first_name + """</td>
            <td>""" + last_name + """</td>
            <td>""" + country_code + """</td>
            <td>""" + str(row['ranking_points']) + """</td>
            </tr>"""
            table += table_row
    table += "</table>"
    
    return HTML(table)
```

We can output ATP Ranking for any date in the databases

```python
# Output ATP Rankings for date 20141229 (29 December of 2014) for the first 100 top players.
output_atp_rankings(2014, 20141229, 100)
```

Let's make initial dig in our available data. As can be seen we have than more 140000 match results for more than 4000 players and 3500 tournaments. Amazing!

```python

total_matches = dict()
won_matches = dict()
lost_matches = dict()
years_range = range(1968, 2015)
tournaments_by_year = dict()
players_by_year = dict()
num_matches = 0
for year in years_range:
    atp_matches = pd.read_csv(ATP_MATCHES_FILENAME + str(year) + '.csv')
    num_matches += len(atp_matches)
    w_players = atp_matches['winner_id'].values
    l_players = atp_matches['loser_id'].values

    for player in w_players:
        if player not in won_matches:
            won_matches[player] = 0
        if player not in total_matches:
            total_matches[player] = 0
        won_matches[player] += 1
        total_matches[player] += 1

    for player in l_players:
        if player not in lost_matches:
            lost_matches[player] = 0
        if player not in total_matches:
            total_matches[player] = 0
        lost_matches[player] += 1
        total_matches[player] += 1

    w_players = set(w_players)
    l_players = set(l_players)
    players = set.union(w_players, l_players)
    tournaments = set(atp_matches['tourney_name'].values)
    tournaments_by_year[year] = tournaments
    players_by_year[year] = players 

print("Total Number of Matches = {0}".format(num_matches))
num_tournaments = [len(tournaments_by_year[year]) for year in years_range]
print("Total Number of Tournaments = {0}".format(np.sum(num_tournaments)))
plt.figure(figsize = (20, 10))
plt.bar(years_range, num_tournaments)
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Number of Tournaments', fontsize = 20)

all_players = set()
num_players = [len(players_by_year[year]) for year in years_range]
for year in years_range:
    all_players = all_players.union(players_by_year[year])
print("Total Number of Players = {0}".format(len(all_players)))
plt.figure(figsize = (20, 10))
plt.bar(years_range, num_players)
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Number of Players', fontsize = 20)

total_matches_sorted = np.array(sorted(total_matches.values(), reverse = True))
won_matches_sorted = np.array(sorted(won_matches.values(), reverse = True))
lost_matches_sorted = np.array(sorted(lost_matches.values(), reverse = True))

match_thresholds = [1, 5, 10, 25, 50, 100, 150, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
total_fraction = list()
lost_fraction = list()
won_fraction = list()
for match_threshold in match_thresholds:
    total_fraction.append(len(np.where(total_matches_sorted >= match_threshold)[0]) / float(len(all_players)))
    won_fraction.append(len(np.where(won_matches_sorted >= match_threshold)[0]) / float(len(all_players)))
    lost_fraction.append(len(np.where(lost_matches_sorted >= match_threshold)[0]) / float(len(all_players)))

plt.figure(figsize = (20, 10))
area = np.pi * 10 **2
alpha = 0.5
plt.hold(True)
plt.scatter(match_thresholds, -np.log10(total_fraction), s = area, color = 'black', label = 'Total', alpha = alpha)
plt.scatter(match_thresholds, -np.log10(won_fraction), s = area, color = 'red', label = 'Won', alpha = alpha)
plt.scatter(match_thresholds, -np.log10(lost_fraction), s = area, color = 'blue', label = 'Lost', alpha = alpha)
plt.grid(True)
plt.legend(loc='upper left')
plt.xlabel('Number of Matches', fontsize = 20)
plt.ylabel('-Log10(Fraction of Players)', fontsize = 20)
```

Okay, it is pretty nice. But let's move to the most interesting part of this analysis: to Pagerank!

```python
# Our weights matrix is just number of loses one player to another. Pretty simple.
def get_head2head(years, surfaces = 'all'):
    weights = dict()
    for year in years:
        atp_matches = pd.read_csv(ATP_MATCHES_FILENAME + str(year) + '.csv')
        match_surfaces = atp_matches['surface'].values
        w_players = atp_matches['winner_id'].values
        l_players = atp_matches['loser_id'].values
        for (w_player, l_player, match_surface) in zip(w_players, l_players, match_surfaces):
            if surfaces == 'all' or match_surface in surfaces:
                if (w_player, l_player) not in weights:
                    weights[(w_player, l_player)] = 0
                    weights[(l_player, w_player)] = 0
                weights[(w_player, l_player)] +=1
            
    return weights

def get_pagerank_rankings(years, surfaces):
    weights = get_head2head(years, surfaces)
    G = nx.DiGraph() 
    for key in weights:
        G.add_weighted_edges_from([(key[1], key[0], weights[key])])
    
    # Use pagerank from networkx library - so easy!
    pagerank = nx.pagerank_numpy(G, alpha = 0.85)
    pagerank_rankings = sorted(pagerank.items(), key=operator.itemgetter(1), reverse = True)
    return pagerank_rankings

# Another output fucntion, now for Pagerank Rankings 
def output_pagerank_rankings(pagerank_rankings, max_num_players):
    atp_players = pd.read_csv(ATP_PLAYERS_FILENAME)
    table = """<table>
    <tr>
    <th>Ranking</th>
    <th>First Name</th>
    <th>Last Name</th>
    <th>Country</th>
    <th>Points</th>
    </tr>
    """
    for (ranking, pagerank_ranking) in enumerate(pagerank_rankings[:max_num_players]):
            player = atp_players[atp_players['player_id'] == pagerank_ranking[0]]
            first_name = player['first_name'].values[0]
            last_name = player['last_name'].values[0]
            country_code = player['country_code'].values[0]
            table_row = """
            <tr>
            <td>""" + str(ranking + 1) + """</td>
            <td>""" + first_name + """</td>
            <td>""" + last_name + """</td>
            <td>""" + country_code + """</td>
            <td>""" + str(pagerank_ranking[1]) + """</td>
            </tr>"""
            table += table_row
    table += "</table>"
    
    return HTML(table)
```

The necessary functions were written, so it is time to find the best ATP player of all time! And it is Jimmy Connors! 

```python
pagerank_rankings = get_pagerank_rankings(range(1968, 2015), 'all')
output_pagerank_rankings(pagerank_rankings, 30)
```
